```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
house_data = pd.read_csv("../preprocessed_data/oneHot_data.csv")
```

```python
house_data = house_data[['T_ctrl', 'T_stp_heat', 'T_out', 'RH_out', 'auxHeat1']]
```

```python
plt.figure(figsize = (20, 15))
ax = sns.barplot(x = house_data.auxHeat1.value_counts().index, y = house_data.auxHeat1.value_counts(normalize = True).values)
ax.set(xlabel = 'Target Variable Values', ylabel = 'Percentage Values', ylim = (0, 1))
```
